```python
destination=[]
normal=[]
activities_combined=[]

# spot_combined=[]

for i in wisata[732:]:
    print (i)
    destination.append(i)
    
    #opening the website
    try:
        driver.get("https://www.tripadvisor.co.id/")
        sleep(3)
    except:
        try:
            driver.get("https://www.tripadvisor.co.id/")
            sleep(5)
        except:
            driver.get("https://www.tripadvisor.co.id/")
            sleep(5)
            
    todo=driver.find_element_by_xpath('//*[@id="lithium-root"]/main/div[1]/div[1]/div/div/div[2]/a/span/span[1]')
    todo.click()
    sleep(2)

    searchtodo=driver.find_element_by_css_selector('body > div._5_xsoLU2 > div > form > input._3qLQ-U8m')
    searchtodo.send_keys(i)
    sleep(2)
    searchtodo.send_keys(Keys.ENTER)
    sleep(2)
    
    try:
        check=driver.find_element_by_xpath('//*[@id="lithium-root"]/main/nav/div/div/div[1]/a[1]/span')
        normal.append(check.text)
        print(check.text)
    except:
        normal.append(np.nan)
    
    try:
        types=driver.find_elements_by_class_name("_3S09qsQh._30GXgBoj")

        jeniss=[]
        for jenis in types:    
            if jenis.text!="":
                jeniss.append(jenis.text)
        activities_combined.append(jeniss)
        print(jeniss)

        sleep(3)
    
    except:
        jeniss.append(np.nan)
```
