```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep, strftime
from random import randint
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\chiri\Purwadhika\Final Project\dataset_indo.csv")

destination=df["Travel Destination"].values
```

```python
df=df.reset_index().drop(["Unnamed: 0", "index"], axis=1)
```

```python
df
```

```python
homepage = "https://www.tripadvisor.co.id/Attractions"
path = r"C:\Users\chiri\Downloads\chromedriver.exe"

driver = webdriver.Chrome(path)
driver.get(homepage)
```

```python
df_destiny=[]
df_linkhotel=[]
df_besthotel=[]
df_linkflight=[]
df_bestflight=[]
```

```python
for destiny in destination:
    df_destiny.append(destiny)
    print(destiny)
    driver.get(homepage)
    sleep(3)
    driver.find_element_by_xpath('//*[@id="component_1"]/div/div/form/input[1]').send_keys(destiny)
    sleep(2)
    driver.find_element_by_class_name('_1dvyiAq4').click()
    sleep(2)
    
    #nyari submenu : Hotel dari list submenu yang ada

    driver.find_elements_by_class_name("_1T4t-FiN")
    tes=driver.find_elements_by_class_name("_1T4t-FiN")
    
    #nyari submenu : Hotel dari list submenu yang ada
    if len(tes)==0:
        print("a")
        df_linkhotel.append(np.nan)
        df_besthotel.append(np.nan)
        df_linkflight.append(np.nan)
        df_bestflight.append(np.nan)
    
    else:
        for i in tes:
            try:
                if (i.text)=="Hotel":
                    try:
                        print(i.get_attribute("href"))
                        df_linkhotel.append(i.get_attribute("href"))
                        i.click()
                        sleep(3)
                    except:
                        df_linkhotel.append(np.nan)
                    
                    try:
                        print(driver.find_elements_by_class_name("price.__resizeWatch")[0].text)
                        df_besthotel.append(driver.find_elements_by_class_name("price.__resizeWatch")[0].text)
                    except:                    
                        df_besthotel.append(np.nan)
                        print("b")

                    #navigasi ke tab penerbangan
                    try:
                        pesawat=driver.find_element_by_id("global-nav-flights")
                        print(pesawat.get_attribute("href"))
                        df_linkflight.append(pesawat.get_attribute("href"))
                        pesawat.click()
                        sleep(5)
                    except:
                        df_linkflight.append(np.nan)
                        print("c")

                    #Cari Penerbangan
                    driver.find_element_by_class_name("_1JOGv2rJ._1_M9wxW9._2oWqCEVy._3Ffza2n4._3fiJJkxX").click()
                    sleep(10)

                    #ambil harga terendah
                    try:
                        print(driver.find_elements_by_class_name("zEpUay0i")[0].text)
                        df_bestflight.append(driver.find_elements_by_class_name("zEpUay0i")[0].text)              
                    except:
                        df_bestflight.append(np.nan)
                        print("d")

            except:
                continue
```

```python
print(len(df_destiny))
print(len(df_linkhotel))
print(len(df_besthotel))
print(len(df_linkflight))
print(len(df_bestflight))
```
