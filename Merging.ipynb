we will first slice the dataframe for only the not-null flights. I will then fill out the null values using the mean of the Flight Best Deal from the available data, with first group by their location. This is why I need to change them into int first so I can calculate the mean.

```python
notnull_flight=df[df["Flight Best Deal"].notnull()]
```

```python
notnull_flight["Flight Best Deal"]=notnull_flight["Flight Best Deal"].astype("int")
```

```python
location_data=notnull_flight.groupby("location").mean()
```

```python
location_data
```

```python
null_flight=df[df["Flight Best Deal"].isnull()]
```

```python
for i in null_flight.index:
    if null_flight["location"][i] in location_data.index:
        df["Flight Best Deal"][i]=location_data.loc[df["location"][i]]["Flight Best Deal"]
```

```python
df["Hotel Best Deal"]=df["Hotel Best Deal"].astype(str)
df["Flight Best Deal"]=df["Flight Best Deal"].astype(str)
```

```python
def rupiah_hotel(x):
    if (len(x["Hotel Best Deal"])<=6)&(len(x["Hotel Best Deal"])>3):
        return x["Hotel Best Deal"][:-3]+"."+x["Hotel Best Deal"][-3:]
    elif len(x["Hotel Best Deal"])>=7:
        return x["Hotel Best Deal"][:-6]+"."+x["Hotel Best Deal"][-6:-3]+"."+x["Hotel Best Deal"][-3:]
    else:
        return x["Hotel Best Deal"]
```

```python
df["Hotel Best Deal"]=df.apply(rupiah_hotel,axis=1)
```

```python
def rupiah_flight(x):
    if (len(x["Flight Best Deal"])<=6)&(len(x["Flight Best Deal"])>3):
        return x["Flight Best Deal"][:-3]+"."+x["Flight Best Deal"][-3:]
    elif len(x["Flight Best Deal"])>=7:
        return x["Flight Best Deal"][:-6]+"."+x["Flight Best Deal"][-6:-3]+"."+x["Flight Best Deal"][-3:]
    else:
        return x["Flight Best Deal"]
```

```python
df["Flight Best Deal"]=df.apply(rupiah_flight,axis=1)
```
